# Método de Euler

```julia
function analytical_solution(T0, g, l, t)
    ω = sqrt(g / l)
    return T0 * cos(ω * t)
end
function euler_method(T0, dT0, g, l, h, num_steps)
    dt = h / num_steps

    T = zeros(num_steps + 1)
    dT = zeros(num_steps + 1)

    T[1] = T0
    dT[1] = dT0

    for i in 1:num_steps
        dT[i + 1] = dT[i] - (g * T[i] / l) * dt
        T[i + 1] = T[i] + dT[i] * dt
    end

    return T, dT
end

function solve_for_discretizations(T0, dT0, g, l, discretizations)
    results = Dict()

    for h in discretizations
        num_steps = Int(round(1.0 / h))
        T, dT = euler_method(T0, dT0, g, l, h, num_steps)
        results[h] = (T, dT)
    end

    return results
end

# Initial conditions
T0 = 0.0
dT0 = 0.5

# Constants
g = 9.81  # acceleration due to gravity (m/s^2)
l = 1.0   # length of the pendulum (m)
discretizations = [0.01, 0.005, 0.002]  # domain discretization values

# Solving the differential equation using the Euler method for multiple discretizations
results = solve_for_discretizations(T0, dT0, g, l, discretizations)

# Printing the results for each discretization
for (h, (T, dT)) in results
    println("Discretization h = $h")
    println("Time (s)\tAngle (rad)\tAngular Velocity (rad/s)")
    num_steps = length(T)
    for i in 1:num_steps
        time = i * h
        println("$time\t\t$(T[i])\t\t$(dT[i])")
    end
    println()
end

function solve_for_discretizations(T0, dT0, g, l, discretizations, total_time)
    results = Dict()

    for h in discretizations
        num_steps = Int(round(total_time / h))
        T, dT = euler_method(T0, dT0, g, l, h, num_steps)
        results[h] = (T, dT)
    end

    return results
end

# Initial conditions
T0 = 0.0
dT0 = 0.5

# Constants
g = 9.81  # acceleration due to gravity (m/s^2)
l = 1.0   # length of the pendulum (m)
discretizations = [0.01, 0.005, 0.002]  # domain discretization values
total_time = 10.0  # Total simulation time (s)

# Solving the differential equation using the Euler method for multiple discretizations
results = solve_for_discretizations(T0, dT0, g, l, discretizations, total_time)

# Analytical solution
analytical_T = [analytical_solution(T0, g, l, i * 0.01) for i in 0:1000]

# Printing the results for each discretization and comparing with analytical solution
for (h, (T, _)) in results
    println("Discretization h = $h")
    println("Time (s)\tNumerical T\tAnalytical T\tError")
    num_steps = length(T)
    for i in 1:num_steps
        time = i+h
        analytical_T_val = analytical_T[i]
        error = abs(T[i] - analytical_T_val)
        println("$time\t\t$(T[i])\t\t$analytical_T_val\t\t$error")
    end
    println()
end
```

a função converge com os menores números de de h que foram analisados nesse momento.

# Runge-Kutta 4ª Ordem

```julia
T0 = 0.0
dT0 = 0.5
function rk4_method(T0, dT0, g, l, h, num_steps)
    
    T = zeros(num_steps + 1)
    dT = zeros(num_steps + 1)

    T[1] = T0
    dT[1] = dT0

    for i in 1:num_steps
        k1_T = dT[i]
        k1_dT = - (g * T[i] / l)

        k2_T = dT[i] + 0.5 * h * k1_dT
        k2_dT = - (g * (T[i] + 0.5 * h * k1_T) / l)

        k3_T = dT[i] + 0.5 * h * k2_dT
        k3_dT = - (g * (T[i] + 0.5 * h * k2_T) / l)

        k4_T = dT[i] + h * k3_dT
        k4_dT = - (g * (T[i] + h * k3_T) / l)

        dT[i + 1] = dT[i] + (h / 6) * (k1_dT + 2 * k2_dT + 2 * k3_dT + k4_dT)
        T[i + 1] = T[i] + (h / 6) * (k1_T + 2 * k2_T + 2 * k3_T + k4_T)
    end

    return T, dT
end

# Constants
g = 9.81  # acceleration due to gravity (m/s^2)
l = 1.0   # length of the pendulum (m)
discretizations = [0.01, 0.005, 0.002]  # domain discretization values
total_time = 10.0  # Total simulation time (s)
function analytical_solution(T0, g, l, t)
    ω = sqrt(g / l)
    return T0 * cos(ω * t)
end

function solve_for_discretizations(T0, dT0, g, l, discretizations, total_time)
    results = Dict()

    for h in discretizations
        num_steps = Int(round(total_time / h))
        T, dT = rk4_method(T0, dT0, g, l, h, num_steps)
        results[h] = (T, dT)
    end

    return results
end
# Solving the differential equation using the 4th order Runge-Kutta method for multiple discretizations
results_rk4 = solve_for_discretizations(T0, dT0, g, l, discretizations, total_time)

# Printing the results for each discretization using Runge-Kutta method and comparing with analytical solution
for (h, (T_rk4, _)) in results_rk4
    println("Discretization h = $h (Runge-Kutta Method)")
    println("Time (s)\tNumerical T\tAnalytical T\tError")
    num_steps = length(T_rk4)
    for i in 1:num_steps
        time = i * h
        analytical_T_val = analytical_solution(T_rk4[i], g, l, time)
        error = abs(T_rk4[i] - analytical_T_val)
        println("$time\t$(T_rk4[i])\t$analytical_T_val\t$error")
    end
    println()
end
```
