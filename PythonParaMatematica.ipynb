```python
import numpy as np

coeficientes = np.array([[3, 2],[1, -3]])
resultados = np.array([7, 1])
solucao = np.linalg.solve(coeficientes, resultados)
print(solucao)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(solucao[0], solucao[1], 'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.title ('Soluo do sistema linear')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(solucao[0], solucao[1], 'ro')  
reta1 = (resultados[0] - coeficientes[0,0] * x) / coeficientes[0,1]
reta2 =  (resultados[1] - coeficientes[1,0] * x) / coeficientes[1,1]
plt.plot(x, reta1, label='3x+2y=7')
plt.plot(x, reta2, label='x-3y=1')
plt.xlabel('x')
plt.ylabel('y')
plt.title ('Soluo do sistema linear')
plt.legend()
plt.grid(True)
plt.show()

```

```python
from scipy.integrate import solve_ivp
def func(y,t,k):
    dydt = -k * y
    return dydt
k=0.3
sol = solve_ivp(func, [0, 10], y0, args=(k,), dense_output=True)
y = sol.sol(t)
plt.plot(t, y.T, label='y(t)')
plt.xlabel('t')
plt.legend()
plt.grid(True)
plt.title('ED Solution')
```
